This notebook takes the names that we generated and renames the cell_table acquisition_id. The new names have been found by Leor

```python
import pandas as pd
import numpy as np
import os, shutil
```

```python
def leap3_4():
    swap_from_leap4_to_leap3 =[14,15,16]
    swap_from_leap3_to_leap4 =[11,12,13]
    #old_name:new_name
    file_2_rename = {'Leap003_'+str(id):'Leap004_'+str(id) for id in swap_from_leap3_to_leap4}|{'Leap004_'+str(id):'Leap003_'+str(id) for id in swap_from_leap4_to_leap3}
    files_2_delete = []    
    return files_2_delete,file_2_rename

def leap17_18():
    files_2_delete = ['Leap018_'+str(x)for x in range(14,25)] 
    return files_2_delete,{}
def leap10_11():
    files_2_delete = ['Leap011_7']
    swap_from_leap11_to_leap10 =['6','8','9','10','11','12','13']
    swap_from_leap10_to_leap11 =['1','2','3','4','5']
    file_2_rename = {'Leap011_'+str(id):'Leap010_'+str(id) for id in swap_from_leap11_to_leap10}|{'Leap010_'+str(id):'Leap011_'+str(id) for id in swap_from_leap10_to_leap11}
    return files_2_delete,file_2_rename
def leap24_25():
    files_2_delete = ['Leap024_'+str(x) for x in range(1,7)]+['Leap025_7']
    return files_2_delete,{}    
def leap40():
    files_2_delete = ['Leap040_'+str(x) for x in [2,3,4,5,7]]
    return files_2_delete,{}
def aggregate_changes():
    functions = [leap3_4,leap17_18,leap10_11,leap24_25,leap40]
    files_2_rename = {}
    files_2_delete = []
    for f in functions:
        a,b = f()
        files_2_delete+=a
        files_2_rename|=b
    return files_2_delete,files_2_rename
files_2_delete,files_2_rename = aggregate_changes()
```

```python
len(files_2_delete),len(files_2_rename)
print(sorted(files_2_delete))
```

Changing cell table

```python
cell_table = pd.read_csv('../segmentation/cell_table_Denoised/cell_table_size_normalized_cell_labels.csv',index_col=0)
cell_table = cell_table[~cell_table.fov.isin(files_2_delete)]
cell_table.loc[:,'fov'] = cell_table.loc[:,'fov'].map(files_2_rename).fillna(cell_table.loc[:,'fov'])
```

```python
cell_table.to_csv('../segmentation/cell_table_Denoised/cell_table_size_normalized_cell_labels_corrected_labels.csv',index = False)
```

Removing Img from Img Denoised

```python
for file in files_2_delete:
    shutil.rmtree('../Img_Denoised/'+file)
```

```python
for old,new in files_2_rename.items():
    os.rename('../Img_Denoised/'+old,'../Img_Denoised/'+new)
```

Removing Img from segmentation

```python
for file in files_2_delete:
    os.remove('../segmentation/deepcell_output_Denoised/'+file+'_whole_cell.tiff')
    os.remove('../segmentation/deepcell_output_Denoised/'+file+'_nuclear.tiff')
    
```

```python
for old,new in files_2_rename.items():
    os.rename('../segmentation/deepcell_output_Denoised/'+old+'_whole_cell.tiff','../segmentation/deepcell_output_Denoised/'+new+'_whole_cell.tiff')
    os.rename('../segmentation/deepcell_output_Denoised/'+old+'_nuclear.tiff','../segmentation/deepcell_output_Denoised/'+new+'_nuclear.tiff')
```
