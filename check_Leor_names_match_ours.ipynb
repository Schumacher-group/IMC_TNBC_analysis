```python
import pandas as pd
import os
import pathlib
import glob
import numpy as np
```

```python
leor = pd.read_excel('../IMC_data/ExtraDocs/imc_metadata.xlsx',dtype={'acquisition_raw_id':str})
```

```python
leor
```

```python
ours = glob.glob('../split_channels_nohpf/*')
ours = pd.Series(ours).map(lambda path: pathlib.Path(path).parts[-1])
ours = pd.DataFrame(ours,columns = ['full_name'])
ours['acquisition_ID'] = ours.full_name.str.split('_',n = 2).str[-1]
ours['Leap_ID'] = ours.full_name.str.split('_' ).str[0].str[:7]
ours = ours[ours['Leap_ID'].str.contains('Leap0')]
ours.shape
```

```python
merge = pd.merge(leor,ours, left_on= ['leap_id','acquisition_raw_id'],right_on= ['Leap_ID','acquisition_ID'],how='inner')
merge.shape
```

```python
ours[~ours.full_name.isin(merge.full_name)].sort_
```
