```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle as pkl

from scipy.stats import mannwhitneyu
```

```python
pretreat = pd.read_csv('pretreat_df.csv')
```

```python
pretreat.columns
```

```python
stat, p = mannwhitneyu(pretreat[pretreat['Response']=='Responder']['enrichment_Neutrophil_Monocyte'],
             pretreat[pretreat['Response']=='Non-Responder']['enrichment_Neutrophil_Monocyte'])
```

```python
pretreat.columns
# list columns starting with 'enrichment_'
enrichment_columns = [col for col in pretreat.columns if col.startswith('enrichment_')]
```

```python
enrichment_columns
pretreat_stats = {}
for col in enrichment_columns:
    stat, p = mannwhitneyu(pretreat[pretreat['Response']=='Responder'][col],
                           pretreat[pretreat['Response']=='Non-Responder'][col])
    pretreat_stats[col] = {'statistic': stat, 'p_value': p}

pretreat_stats_df = pd.DataFrame(pretreat_stats).T
```

```python
pretreat_stats_df
```

```python
pretreat_stats_df.to_csv('pretreat_stats.csv')
```

```python
nonresponder = pd.read_csv('nonresponder_df.csv')
```

```python
enrichment_columns = [col for col in nonresponder.columns if col.startswith('enrichment_')]
```

```python
nonresponder_stats = {}
for col in enrichment_columns:
    stat, p = mannwhitneyu(nonresponder[nonresponder['Sample_Type_(pre/post treatment)']=='pre'][col],
                           nonresponder[nonresponder['Sample_Type_(pre/post treatment)']=='post'][col])
    nonresponder_stats[col] = {'statistic': stat, 'p_value': p}

nonresponder_stats_df = pd.DataFrame(nonresponder_stats).T
```
